```python
# =====================================================
# IMPORT LIBRARIES
# =====================================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

# Display Settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

print("Libraries Imported Successfully")
```

```python
# =====================================================
# LOAD MAIN DATASET
# =====================================================

country_year = pd.read_csv("data/country_year_indicators.csv")

print("Dataset Loaded Successfully")
```

```python
# =====================================================
# DATASET SHAPE
# =====================================================

print("Rows:", country_year.shape[0])
print("Columns:", country_year.shape[1])

country_year.shape
```

```python
# =====================================================
# FIRST LOOK AT DATA
# =====================================================

country_year.head()
```

```python
# =====================================================
# COLUMN LIST
# =====================================================

country_year.columns.tolist()
```

```python
# =====================================================
# DATASET STRUCTURE
# =====================================================

country_year.info()
```

```python
# =====================================================
# UNIQUE VALUES IN CATEGORICAL COLUMNS
# =====================================================

categorical_cols = [
    'country_code',
    'country_name',
    'iso3',
    'region',
    'income_group'
]

for col in categorical_cols:
    print(f"\n{col}")
    print("-" * 40)
    print("Unique Values:", country_year[col].nunique())
```

```python
# =====================================================
# YEAR COVERAGE
# =====================================================

print("Minimum Year:", country_year['year'].min())
print("Maximum Year:", country_year['year'].max())

print("\nTotal Years:",
      country_year['year'].nunique())
```

```python
# =====================================================
# DATASET SUMMARY
# =====================================================

summary = pd.DataFrame({
    "Rows": [country_year.shape[0]],
    "Columns": [country_year.shape[1]],
    "Countries": [country_year['country_name'].nunique()],
    "Regions": [country_year['region'].nunique()],
    "Income Groups": [country_year['income_group'].nunique()],
    "Year Start": [country_year['year'].min()],
    "Year End": [country_year['year'].max()]
})

summary
```
