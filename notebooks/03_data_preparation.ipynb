```python
# =====================================================
# YEAR COVERAGE COMPARISON
# =====================================================

print("Country-Year Dataset")

print(
    country_year['year'].min(),
    "-",
    country_year['year'].max()
)

print("\nEconomic Stress Dataset")

print(
    economic_stress['year'].min(),
    "-",
    economic_stress['year'].max()
)
```

```python
# =====================================================
# COUNTRY-YEAR KEY VALIDATION
# =====================================================

country_year_keys = set(
    zip(
        country_year['country_code'],
        country_year['year']
    )
)

economic_stress_keys = set(
    zip(
        economic_stress['country_code'],
        economic_stress['year']
    )
)

print(
    "Keys Missing in Economic Stress:",
    len(country_year_keys - economic_stress_keys)
)

print(
    "Extra Keys in Economic Stress:",
    len(economic_stress_keys - country_year_keys)
)
```
